```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing dataset

```python
dataset = pd.read_csv('drive/My Drive/Machine Learning A_to_Z/Linear Regression/Position_Salaries.csv')
dataset
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:, 2].values
```

#Fitting Linear regression to dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X , y)
```

# fitting polynomial regression to dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly , y)
```

# Visualizing linear regression results

```python
plt.scatter(X , y, color='red')
plt.plot(X, lin_reg.predict(X) , color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# visualizing polynomial regression results

```python
plt.scatter(X , y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly) , color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting new result with linea regression

```python
lin_reg.predict(np.array([[6.5]]))
```

# Predicting new result with polynomial regression

```python
lin_reg_2.predict( poly_reg.transform(np.array([[6.5]])))
```
